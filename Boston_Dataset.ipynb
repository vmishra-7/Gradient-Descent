```python
import numpy as np
import pandas as pd
boston = np.genfromtxt(r"training_boston_x_y_train.csv", delimiter = ',')
X_Test = np.genfromtxt(r"test_boston_x_test.csv", delimiter = ',')
boston.shape, X_Test.shape
```

```python
X = boston[:, 0:13]
Y = boston[:, 13]
```

```python
##For X_Train

## Squared pairs
for i in range(13) :
    col = (X[:, i] ** 2).reshape(-1, 1)
    X = np.append(X, col, axis = 1)
    
## Cross pairs
for i in range(12) :
    for j in range(i + 1, 13) :
        col = (X[:, i] * X[:, j]).reshape(-1, 1)
        X = np.append(X, col, axis = 1)

##For X_Test

## Squared pairs
for i in range(13) :
    col = (X_Test[:, i] ** 2).reshape(-1, 1)
    X_Test = np.append(X_Test, col, axis = 1)
    
## Cross pairs
for i in range(12) :
    for j in range(i + 1, 13) :
        col = (X_Test[:, i] * X_Test[:, j]).reshape(-1, 1)
        X_Test = np.append(X_Test, col, axis = 1)
```

```python
def step_gradient(X, Y, learning_rate, m, c) :
    slope_m = np.zeros(X.shape[1], dtype = float)
    slope_c = 0
    N = len(Y)
    for i in range(N) :
        for j in range(len(slope_m)) :
            slope_m[j] += (-2 / N) * (Y[i] - (m * X[i]).sum() - c) * X[i, j]
        slope_c += (-2 / N) * (Y[i] - (m * X[i]).sum() - c)
    new_m = m - (learning_rate * slope_m)
    new_c = c - (learning_rate * slope_c)
    return new_m, new_c
```

```python
def cost(X, Y, m, c) :
    total_cost = 0
    N = len(Y)
    for i in range(N) :
        total_cost += (1 / N) * ((Y[i] - (m * X[i]).sum() - c) **2)   
    return total_cost
```

```python
import numpy as np
def gd(X, Y, learning_rate, num_iterations) :
    m = np.zeros(X.shape[1], dtype = float)
    c = 0
    for i in range(num_iterations) :
        m, c = step_gradient(X, Y, learning_rate, m, c)
        print(cost(X, Y, m, c))
    return m, c
```

```python
def run() :
    learning_rate = 0.01
    num_iterations = 1000
    m, c = gd(X, Y, learning_rate, num_iterations)
    print(m, c)
    Y_pred = predict(X_Test, m, c)
    np.savetxt('y_pred.csv', Y_pred, fmt='%.12f', delimiter=',')
```

```python
def predict(X_Test, m, c) :
    return np.dot(X_Test, m) + c
```
